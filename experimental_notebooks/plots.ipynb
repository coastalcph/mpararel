```python
import sys
sys.path.append('/home/wsr217/mpararel')
import matplotlib.pyplot as plt
from dataset.create_mpararel import get_languages_to_tuples_count
from mpararel_utils import read_mpararel_templates
import collections
import numpy as np
import seaborn as sns
```

```python
mpararel = read_mpararel_templates("../data/mpararel_reviewed_with_tag")
language_relation_tuples_count = get_languages_to_tuples_count(
    "../data/mpararel_with_mlama_zh_corrected/tuples")
```

```python
tuples_count = collections.defaultdict(dict)
for (language, relation, count) in language_relation_tuples_count:
    tuples_count[language][relation+".jsonl"] = count
```

```python
data = []
for language in mpararel.keys():
    phrases_count = sum(
        [len(mpararel[language][relation])*tuples_count[language][relation]
         for relation in mpararel[language].keys()])
    data.append([language, phrases_count])
```

```python
data.sort(key=lambda x: x[1], reverse=True)
data = np.array(data)
```

```python
sns.set_theme(palette=sns.color_palette('colorblind'))
plt.figure(figsize=(11,4), dpi=150)
plt.bar(data[:, 0], data[:, 1].astype(int))
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.xticks(rotation=70)
plt.xlabel("Languages")
plt.title("Total phrases")
plt.show()
```
